```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
pd.set_option('future.no_silent_downcasting', True)
```

# Reading the Data

```python
data = pd.read_csv('cleaned_churn_data.csv')
data.head()
```

# Analysing the Churn column

## Variable Statistics

```python
data.Churn.describe()
```

We can see by the statistics that we have a slight imbalance on the values, with more examples of False values.

To make it easier to plot the distribution let's replace 1's and 0's for True and False

```python
aux = data.copy()
aux['Churn'] = aux['Churn'].replace({1 : True, 0 : False})
aux['Churn'] = aux['Churn'].infer_objects(copy=False)
```

```python
fig = px.histogram(aux, x = 'Churn')

fig.update_layout(yaxis_title = 'Count')

fig.update_xaxes(
    tickvals=['false', 'true'],
    ticktext=['False', 'True']  
)

fig.show()
```
